```python
#import dependencies
import pandas as pd
import os
import datetime
```

```python
# define filepath
winners_filepath = "Resources/Resources_nominiees and winners/Winners__2010-2022_All Categories.csv"
```

```python
# read the file
winners_df = pd.read_csv(winners_filepath)
```

```python
#split , clean and define new columns
winners_df['Comedy (Network)'] = winners_df['Comedy'].str.split('(').str[1]
winners_df['Comedy (Network)'] = winners_df['Comedy (Network)'].str.split(')').str[0]
winners_df['Comedy'] = winners_df['Comedy'].str.split('(').str[0]
winners_df['Drama (Network)']  = winners_df['Drama'].str.split('(').str[1]
winners_df['Drama'] = winners_df['Drama'].str.split('(').str[0]
winners_df['Drama (Network)'] = winners_df['Drama (Network)'].str.split(')').str[0]
```

```python
#save the dataframe
os.makedirs('Resources', exist_ok=True)  
winners_df = winners_df.to_csv('Resources/Winners_cleaned_2010-2022.csv') 
```
